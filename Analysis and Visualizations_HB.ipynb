```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

total_counts = df.groupby('hotel').size().reset_index(name='total_reservations')

cancellation_counts = df.groupby(['hotel', 'is_canceled']).size().reset_index(name='count')

merged_counts = pd.merge(cancellation_counts, total_counts, on='hotel')

merged_counts['percentage'] = (merged_counts['count'] / merged_counts['total_reservations']) * 100

plt.figure(figsize=(8, 4))

ax1 = sns.barplot(x='hotel', y='percentage', hue='is_canceled', data=merged_counts, palette='Blues')

handles, labels = ax1.get_legend_handles_labels()  # Get handles and labels for the current legend

ax1.legend(handles, ['Not Canceled', 'Canceled'], title='Reservation Status', bbox_to_anchor=(1, 1))

plt.title('Reservation Status by Hotel (Percentage)', size=20)
plt.xlabel('Hotel')
plt.ylabel('Percentage of Reservations')

for index, row in merged_counts.iterrows():
    ax1.text(row['hotel'], row['percentage'] + 1, f"{row['count']} ({row['percentage']:.1f}%)", 
             color='black', ha='center')


plt.tight_layout()
plt.show()


for hotel in merged_counts['hotel'].unique():
    total_bookings = total_counts.loc[total_counts['hotel'] == hotel, 'total_reservations'].values[0]
    cancellations = cancellation_counts.loc[(cancellation_counts['hotel'] == hotel) & (cancellation_counts['is_canceled'] == 1), 'count'].values[0]
    cancellation_percentage = (cancellations / total_bookings) * 100
    
    print(f"Hotel: {hotel}")
    print(f"Total Bookings: {total_bookings}")
    print(f"Cancellations: {cancellations}")
    print(f"Cancellation Percentage: {cancellation_percentage:.2f}%\n")
```

# Resrvation status per month for city hotel and resort hotel

```python
df['reservation_status'] = df['is_canceled'].map({0: 'Not Canceled', 1: 'Canceled'})

df['month'] = pd.to_datetime(df['reservation_status_date']).dt.month

fig, axes = plt.subplots(1, 2, figsize=(18, 8), sharey=True)

city_counts = df[df['hotel'] == 'City Hotel'].groupby(['month', 'reservation_status']).size().unstack(fill_value=0)

city_counts = city_counts.reset_index()

city_counts_melted = city_counts.melt(id_vars='month', value_vars=['Canceled', 'Not Canceled'])

sns.barplot(x='month', y='value', hue='reservation_status', data=city_counts_melted, ax=axes[0], palette='Blues')
axes[0].set_title('City Hotel: Reservation Status per Month', fontsize=18)
axes[0].set_xlabel('Month', fontsize=15)
axes[0].set_ylabel('Number of Reservations', fontsize=15)


resort_counts = df[df['hotel'] == 'Resort Hotel'].groupby(['month', 'reservation_status']).size().unstack(fill_value=0)

resort_counts = resort_counts.reset_index()


resort_counts_melted = resort_counts.melt(id_vars='month', value_vars=['Canceled', 'Not Canceled'])

sns.barplot(x='month', y='value', hue='reservation_status', data=resort_counts_melted, ax=axes[1], palette='Greens')
axes[1].set_title('Resort Hotel: Reservation Status per Month', fontsize=18)
axes[1].set_xlabel('Month', fontsize=15)
axes[1].set_ylabel('')

fig.suptitle('Reservation Status per Month for City and Resort Hotels', fontsize=22)


axes[0].set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
axes[1].set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])


plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()
```

#  Market Segment cancellations Percentage By Both Hotel Types

```python
df['reservation_status'] = df['is_canceled'].map({0: 'Not Canceled', 1: 'Canceled'})

# Calculate total reservations and total cancellations for each market segment by hotel type
cancellation_stats = df.groupby(['hotel', 'market_segment']).agg(
    total_reservations=('is_canceled', 'size'),
    total_cancellations=('is_canceled', 'sum')
).reset_index()

# Calculate cancellation percentage
cancellation_stats['cancellation_percentage'] = (cancellation_stats['total_cancellations'] / cancellation_stats['total_reservations']) * 100

# Set up the figure for the bar plot
plt.figure(figsize=(12, 6))

# Create the bar plot
sns.barplot(x='market_segment', y='cancellation_percentage', hue='hotel', data=cancellation_stats, palette='Set2')

# Set the titles and labels
plt.title('Cancellation Percentage by Market Segment and Hotel Type', fontsize=16)
plt.ylabel('Cancellation Percentage (%)', fontsize=14)
plt.xlabel('Market Segment', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Hotel Type')

# Adjust layout to make it look better
plt.tight_layout()
plt.show()

```

#  Is Lead time  the factor for the cancellations through Market Segment in PRT country?

```python
import matplotlib.pyplot as plt
import seaborn as sns

prt_cancelled_data = cancelled_data[cancelled_data['country'] == 'PRT']

mean_lead_time = prt_cancelled_data.groupby('market_segment')['lead_time'].mean().reset_index()

plt.figure(figsize=(12, 6))


sns.barplot(data=mean_lead_time, x='market_segment', y='lead_time', palette='Set2')

plt.title('Average Lead Time for Cancellations by Market Segment in Portugal (PRT)', fontsize=16)
plt.xlabel('Market Segment', fontsize=14)
plt.ylabel('Average Lead Time (Days)', fontsize=14)

for index, row in mean_lead_time.iterrows():
    plt.text(index, row['lead_time'] + 1, f'{row["lead_time"]:.1f}', 
             color='black', ha='center')

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
