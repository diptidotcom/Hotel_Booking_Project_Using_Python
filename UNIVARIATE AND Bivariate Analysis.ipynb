# 4.HeatMap(Cat-Cat)

```python
pd.crosstab(df['hotel'],df['country'])
```

```python
sns.heatmap(pd.crosstab(df['hotel'],df['country']),annot=True,cmap='coolwarm')
```

```python
import seaborn as sns
import pandas as pd


hotel_country_crosstab = pd.crosstab(df['hotel'], df['country'])

top_5_countries = hotel_country_crosstab.sum().nlargest(5).index

hotel_country_top5 = hotel_country_crosstab[top_5_countries]

sns.heatmap(hotel_country_top5, annot=True, cmap='coolwarm')
```
